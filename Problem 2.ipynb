Problem 2:
Even after trying for so many times, your friend’s startup could not find the investment. So you decided to take this matter in your hand and try to find the list of investors who probably can invest in your friend’s startup. Your list will increase the chance of your friend startup getting some initial investment by contacting these investors. Find the top 5 investors who have invested maximum number of times (consider repeat investments in one company also). In a startup, multiple investors might have invested. So consider each investor for that startup. Ignore undisclosed investors.

```python
# Importing Pandas, Numpy, Matplotlib Libraries
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
from collections import Counter

#Reading CSV file
df=pd.read_csv("startup_funding.csv",encoding = 'utf-8')

#Removeing Non Zero values
df.dropna(subset=['Investors Name'],inplace=True)

#ignoring the values of investors whose name is given as "Undisclosed Investors"
df=df[df['Investors Name']!="Undisclosed Investors"]
df=df[df['Investors Name']!="undisclosed Investors"]
df=df[df['Investors Name']!="undisclosed investors"]
df=df[df['Investors Name']!="undisclosed investor"]
df=df[df['Investors Name']!=""]
names=df['Investors Name']
l=[]
for i in names:
    if i!="":
        #seperating Investor Names by seperator ","
        temp=i.split(",")
        for j in temp:
            l.append(j.strip())

#copying the list values to Dictionary
d={}
for i in l:
    d[i]=d.get(i,0)+1

#Using Counter to count the number of occurance of investor names
k=Counter(d)
ans=dict(k.most_common(10))
labels=[]
values=[]
count=0
for i in ans:
    if i!="" and count<5:
        print(i,ans[i])
        labels.append(i)
        values.append(ans[i])
        count=count+1

#Ploting Pie Chart
plt.pie(values,labels=labels,autopct="%.2f%%",counterclock=False,startangle=90)
plt.title("Investor Names vs Percentage of funding")
plt.show()
```
